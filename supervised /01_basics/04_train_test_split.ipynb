**Import bibliotek**

```python
import numpy as np
import pandas as pd
import sklearn
import seaborn as sns

np.random.seed(42) #"ziarno losowe"
np.set_printoptions(edgeitems=30, linewidth = 100000,
                    formatter = dict(float=lambda x:f'{x:.2f}'))
sns.set()
sklearn.__version__
```

**Załadowanie danych**

```python
from sklearn.datasets import load_iris

raw_data = load_iris()
raw_data_copy = raw_data.copy()
raw_data_copy.keys()
```

```python
print(raw_data_copy['DESCR'])
```

**Przygotowanie danych**

```python
data = raw_data_copy['data']
target = raw_data_copy['target']

print(f'{data[:5]}\n')
print(target[:5])   #zmienna docelowa poniżej wyświetla się w klasie 0
```

```python
#połączenie atrybutów ze zmienną docelową
all_data = np.c_[data, target] #połączenie danych data i target
all_data[:5]
```

```python
#budowa obiektu DataFrame
df=pd.DataFrame(data=all_data, columns=raw_data.feature_names + ['target']) #dodanie nazwy dla kolumny target
df.head()
```

```python
df.tail() #wyswietlenie 5 ostatnich wierszy
```

```python
df.info()
```

```python
df.describe() #proste statystyki, wszystkie dane są numeryczne wiac są wyświetlone
```

```python
df.describe().T.apply(lambda x:round(x, 2)) #transponowanie i zaokrąglanie
```

```python
df.target.value_counts() #znalezienie rozkładu zmiennej target
```

```python
df.target.value_counts().plot(kind='pie') #wykres kołowy
```

```python
data = df.copy()
target = data.pop('target') #pop-wyrywa kolumnę target z data, przypisuje ją do zmiennej target, a data nie ma juz target
data.head()
```

```python
target.head()
```

**Podział danych na zbiór treningowy i testowy - iris data** - za pomocą train_test_split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42) #42 to wartość ziarna losowego, czyli będzie ciągle to samo losowanie

print(f'X_train shape {X_train.shape}') #wydruk kształtu-rozmiaru danych
print(f'y_train shape {y_train.shape}')
print(f'X_test shape {X_test.shape}')
print(f'y_test shape {y_test.shape}')
print(f'\nTest ratio: {len(X_test) / len(data):.2f}') #wydrukowanie procentowo testowego
print(f'\ny_train:\n {y_train.value_counts()}') #rozkład zbioru treningowego
print(f'\ny_test: \n {y_test.value_counts()}')
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.3, random_state=42) #teraz 30% danych to dane testowe (dzięki test_size)

print(f'X_train shape {X_train.shape}')
print(f'y_train shape {y_train.shape}')
print(f'X_test shape {X_test.shape}')
print(f'y_test shape {y_test.shape}')
print(f'\nTest ratio: {len(X_test) / len(data):.2f}')
print(f'\ny_train:\n {y_train.value_counts()}')
print(f'\ny_test: \n {y_test.value_counts()}')
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, train_size=0.9, random_state=42)  #rozmiar danych treningowych ustalony na 90%

print(f'X_train shape {X_train.shape}')
print(f'y_train shape {y_train.shape}')
print(f'X_test shape {X_test.shape}')
print(f'y_test shape {y_test.shape}')
print(f'\nTest ratio: {len(X_test) / len(data):.2f}')
print(f'\ny_train:\n {y_train.value_counts()}')
print(f'\ny_test: \n {y_test.value_counts()}')
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42, test_size=0.1, stratify=target)  #równy podział ze względu na zmienną docelową

print(f'X_train shape {X_train.shape}')
print(f'y_train shape {y_train.shape}')
print(f'X_test shape {X_test.shape}')
print(f'y_test shape {y_test.shape}')
print(f'\nTest ratio: {len(X_test) / len(data):.2f}')
print(f'\ny_train:\n {y_train.value_counts()}')
print(f'\ny_test: \n {y_test.value_counts()}')
```

**Podział danych na zbiór treningowy i testowy** - breast cancer data

```python
from sklearn.datasets import load_breast_cancer

raw_data = load_breast_cancer()
raw_data_copy = raw_data.copy()
raw_data_copy.keys()
```

```python
print(raw_data_copy['DESCR'])
```

```python
data = raw_data_copy ['data']
target = raw_data_copy['target']

print(f'{data[:5]}\n')
print(target[:5])
```

```python
all_data = np.c_[data, target]
all_data[:5]
```

```python
df = pd.DataFrame(data=all_data, columns = list(raw_data['feature_names']) + ['target'])
df.head()
```

```python
df.target.value_counts()
```

Rozkład nie jest równy, więc wykonujemy podział warstwowy:

```python
data = df.copy()
target = data.pop('target')
data.head()
```

```python
target.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.25, random_state=40)

print(f'X_train shape {X_train.shape}')
print(f'y_train shape {y_train.shape}')
print(f'X_test shape {X_test.shape}')
print(f'y_test shape {y_test.shape}')
print(f'\nTest ratio: {len(X_test) / len(data):.2f}')
print(f'\ntarget: \n{target.value_counts() / len(target)}')
print(f'\ny_train:\n {y_train.value_counts()}')
print(f'\ny_test: \n {y_test.value_counts()}')
```

**Równy podział ze względu na zmienną docelową**

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.25, random_state=40, stratify = target)

print(f'X_train shape {X_train.shape}')
print(f'y_train shape {y_train.shape}')
print(f'X_test shape {X_test.shape}')
print(f'y_test shape {y_test.shape}')
print(f'\nTest ratio: {len(X_test) / len(data):.2f}')
print(f'\ntarget: \n{target.value_counts() / len(target)}')
print(f'\ny_train:\n {y_train.value_counts()}')
print(f'\ny_test: \n {y_test.value_counts()}')
```
