Importowanie biblioteki pandas

```python
import numpy as np
import pandas as pd
import sklearn

sklearn.__version__
```

Wygenerowanie danych

```python
data = {
    'size': ['XL','L','M','L','M'],
    'color': ['red','green','blue','green','red'],
    'gender': ['female', 'male', 'male', 'female', 'female'],
    'price':[199.0, 89.0, 99.0, 129.0, 79.0],
    'weight': [500, 450, 300, 380, 410],
    'bought': ['yes', 'no', 'yes', 'no', 'yes']
}

df_raw = pd.DataFrame(data=data)
df_raw
```

Utworzenie kopii danych

```python
df = df_raw.copy()
df.info()
```

Przetwarzamy zmienne obiect na postaci kategoryczne.

Zmiana typu danych (astype) i wstępna eksploracja:

```python
for col in ['size', 'color', 'gender', 'bought']:
  df[col] = df[col].astype('category')

df['weight'] = df['weight'].astype('float')

df.info()
```

```python
df.describe()
```

transponowanie powyższej tabeli

```python
df.describe().T
```

wyświetlenie statystyk zmiennych category

```python
df.describe(include=['category']).T
```

Przygotowanie danych do uczenia maszynowego

LabelEncoder - klasa pozwalająca nam dane z "bought" przeformatować na system 0-1

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
le.fit(df['bought'])
le.transform(df['bought'])
```

le.fit i le.transform możemy połaczyć do le.fit_transform

```python
le.fit_transform (df['bought'])
```

Pokazanie, jak przebiega mapowanie:

```python
le.classes_
```

Przypisanie na stałe powyższej zmiany (mapowania) do obiektu:

```python
df['bought']=le.fit_transform(df['bought'])
df
```

Jeśli byśmy chcieli wrócić do oryginalnej postaci:

```python
le.inverse_transform(df['bought'])
```

```python
df['bought']=le.inverse_transform(df['bought'])
df
```

OneHotEncoder - kodowanie zmiennych opisujących/objaśniająch

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse=False)
encoder.fit(df[['size']])
```

```python
encoder.transform (df[['size']])
```

```python
encoder.categories_
```

drop='first' - usunięcie pierwszej z kolumn

```python
encoder = OneHotEncoder(drop='first', sparse=False)
encoder.fit(df[['size']])
encoder.transform(df[['size']])
```

Standaryzacja - StandardScaler

std() - pandas nieobciążony

std() - numpy obciążony

```python
print(f"{df['price']}\n")
print(f"Średnia: {df['price'].mean()}")
print(f"Odchylenie standardowe: {df['price'].std():.4f}")
```

```python
(df['price'] - df['price'].mean()) / df['price'].std()
```

Możemy również zbudować funkcję:

```python
def standarize(x):
  return (x - x.mean()) / x.std()

standarize(df['price'])
```

Korzystanie gotowej funkcji scale - wartości się różnią ze względu na wartośc estymatora obciążonego/nieobciążonego wariancji

```python
from sklearn.preprocessing import scale

scale (df['price'])
```

Lepiej działa klasa StandardScaler:

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(df[['price']])
scaler.transform(df[['price']])
```

Przekształcanie - standaryzacja jednocześnie wszystkich kolumn numerycznych

```python
scaler = StandardScaler()
df[['price','weight']] = scaler.fit_transform(df[['price','weight']])
df
```
